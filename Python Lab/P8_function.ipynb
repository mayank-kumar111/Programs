```python
# in python function is object. a, b, c, d, x are parameters. they are local variable. they are only exist in function.
def add(a, b) :
    return a + b
```

```python
def product(c, d):
    return c * d
```

```python
def square(x):
    return x * x
```

```python
print(add)
p = add
p(10, 20)
print(f" Type of add is {type(add)} and type of p is {type(p)}")
result = square(10)
print(f" square of 10 is {result}")
```

```python
# Types of arguments
# 1. positional argument, order matters
r = product(10, 20)
print(f" product of 10 and 20 is {r}")

# 2. keyword argument, order does not matter also know as named argument 
# note: positional argument and keyword argument can be used together but positional argument should be before keyword argument
r = product(d=20, c=10)
print(f" product of 10 and 20 is {r}")

# 3. optional/default argument, if we do not pass value for parameter then default value will be used
def power(x, n=2):
    return x ** n

# 4. variable length argument, if we do not know how many arguments will be passed to function then we can use variable length argument. *args is used to pass variable length argument to function. it will be treated as tuple in function. **kwargs is used to pass variable length keyword argument to function. it will be treated as dictionary in function.
# note: we can use *args and **kwargs together but *args should be before **kwargs
def fun(*args):
    print(f"Variable length argument")
    print(f" args is {args} and type of args is {type(args)}")
    t= args
    sum = 0
    for i in t:
        sum = sum + i
    print(f" sum of all arguments is {sum}")

fun(10, 20, 30)

def fun1(**kwargs):
    print(f"Variable length keyword argument")
    print(f" kwargs is {kwargs} and type of kwargs is {type(kwargs)}")

    for key, value in kwargs.items():
        print(f" key is {key} and value is {value}")

fun1(Name="Mayank",age="20",dep="IT")

fun1(a=10, b=20, c=30)
```

```python
# returning multiple values from function
def add_product(x=10, y=20):
    return x + y, x * y

# tuple packing
t = add_product()
print(f" Type of t is {type(t)}")
print(f" tuple returned from add_product is {t}")

# tuple unpacking
sum,prod = t
print(f" sum is {sum} and product is {prod}")

# or we can do it in single line tuple packing and unpacking
sum, prod = add_product()
print(f" sum is {sum} and product is {prod}")
```
