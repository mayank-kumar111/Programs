```python
# Student Marks Calculation (Arithmetic Operations)
my_name = "Mayank Kumar"
mid_term_1_marks = 25
mid_term_2_marks = 30
total_ques = 50
attempted_ques = 45

total_mid_marks = mid_term_1_marks + mid_term_2_marks
print(f"{my_name} score total marks = {total_mid_marks}")

#Total Leftover question
left_over_ques = total_ques - attempted_ques
print(f"Left Over Ques are:{left_over_ques}")

per_ques_marks = 2
max_marks = total_ques*per_ques_marks
print(f"Maximum marks in paper is:{max_marks}")

```

```python
# Check Even Odd (Function) 
def check_even_odd(n):
    if n%2 ==0:
        print(f"{n} is even")
    else:
        print(f"{n} is odd")  
check_even_odd(20) 
check_even_odd(15)
check_even_odd(0)
check_even_odd(-5)
```

```python
# TUPLE DATA TYPE
#tuple creation
t= ("Mayank Kumar",22,"SKIT,Jaipur")
print(f"Type of t is:{type(t)}")

#tuple unpacking
name, age, address = ("Mayank Kumar",22,"SKIT,Jaipur")
print(f"{name}--{age}--{address}")

#accessing tuple members
print(f"1st member of tuple:{t[0]}...Second mmber is:{t[1]}..Third member is:{t[2]}")
print(f"Mayank Kumar" in t)

#iterate through tuple
numbers_tuple = (10,20,30,40,50)
sum = 0
max = 0
for number in numbers_tuple:
    sum = sum + number
    if number > max:
        max = number
print(f"Sum of all numbers is: {sum}")
print(f"maximum among all number:{max}")

# Tuple Slicing and Methods
weekly_temperatures = (30,32,31,29,28,27,26)
weekend_temperatures = weekly_temperatures[5:7]
print(f"Weekend temp are: temp are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[0:7:2]
print(f"All week temperature are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[0:7:1]
print(f"All week temperature are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[::-1]
print(f"All week temperature are in reverse:{weekend_temperatures}")

count_30=weekly_temperatures.count(30)

print(f"Count of 30:{count_30}")
```

```python
# Armstrong Number Check
number =int(input("Please enter the number"))
k=number//10
r=number%10
if (k**1+r**2==number):
    print("True")
else:
    print("False")
```

```python
# Diserian number Check
number =int(input("Please enter the number"))
sum=0
temp=number
while temp>0:
    digit = temp%10
    sum += digit**3
    temp//=10
if sum == number:
    print(f"{number} is Diserian number")
else:
    print(f"{number} is not Diserian number")
```
