```python
'''Goal: find high-value transactions.
The Task: A manager wants to see a list of all purchases that were over $2000 to send "Thank You" notes to those customers.
The Data: orders = [1200, 4500, 800, 2100, 5000, 150]
customers = ["Ram", "Mayank", "Piyush", "Mohit", "Pratik", "Satyarth"]'''

# Without lambda function and filter and zip
orders = [1200, 4500, 800, 2100, 5000, 150]
customers = ["Ram", "Mayank", "Piyush", "Mohit", "Pratik", "Satyarth"]
'''high_value_customers = []
for order, customer in zip(orders, customers):
    if order > 2000:
        high_value_customers.append(customer)
print(f"High value customers are: {high_value_customers}")'''

high_value_customers = list(filter(lambda customer: orders[customers.index(customer)] > 2000, customers))
print(f"High value customers are: {high_value_customers}")
```

```python
'''Square of side 'N'
Problem Description: You are given an integer n. Your task is to return a square pattern of size n x n made up of the character '*', represented as a list of strings.

Input Parameters:
n (int): The size of the square (number of rows and columns).
Output:
A list of strings where each string is a row of n characters.
Example:
Input: 3
Output: ['***', '***', '***']
Input: 5
Output: ['*****', '*****', '*****', '*****', '*****']'''

def string_list(n):
    list_of_strings = []
    for i in range(n):
        list_of_strings.append('*' * n)
    return list_of_strings

print(string_list(3))
print(string_list(5))
    
```

```python
'''Hollow Square of side 'N'
Problem Description:

You are given an integer n. Your task is to return a hollow square pattern of size n x n made up of the character '', represented as a list of strings. The hollow square has '' on the border, and spaces ' ' in the middle (except for side lengths of 1 and 2).

Input Parameters:
n (int): The size of the square (number of rows and columns).

Output:
A list of strings where each string is a row of n characters, representing a hollow square.

Example:
Input: 3
Output: ['***', '* *', '***']

Input: 5
Output: ['*****', '*   *', '*   *', '*   *', '*****']'''

def hollow_string_list(n):
    list_of_strings = []
    for i in range(n):
        if i == 0 or i == n - 1:
            list_of_strings.append('*' * n)
        else:
            list_of_strings.append('*' + ' ' * (n - 2) + '*')
    return list_of_strings
print(hollow_string_list(3))
print(hollow_string_list(5))
```
