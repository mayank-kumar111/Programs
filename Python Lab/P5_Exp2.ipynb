```python
# Write a program to compute distance between two points taking input from the user
import math
def compute_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

x1,y1,x2,y2 = map(float, input("Enter coordinates of two points (x1 y1 x2 y2): ").split())
# x,y = [int(x) for x in input("Enter Two numbers separated by a comma: ").split(",")]
distance = compute_distance(x1, y1, x2, y2)
print(f"The distance between the points ({x1}, {y1}) and ({x2}, {y2}) is {distance:.2f}")
```

```python
# Write a program add.py that takes 2 numbers as command line arguments and prints its sum.
from sys import argv
cmd_argument = argv[1:]
sum =0
for number in cmd_argument:
    sum += int(number)
print(f"Sum of command line argument is : {sum}")

# For Run start terminal python file_name.py argument
# Eg: python P5_Exp2.py 10 20
```
