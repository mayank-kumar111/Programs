```python
# find s algorithm
import pandas as pd
data = pd.read_csv("Play_tennis_dataset.csv")

X = data.iloc[:, :-1].values   
y = data.iloc[:, -1].values
print("The data is: \n",X)
print("The target is:",y)

hypo = ["0"] * len(X[0])
print("The initial hypothesis is: \n",hypo)

for i in range(len(X)):              
    if y[i] == "Yes":                
        for j in range(len(hypo)):  
            if hypo[j] == "0":
                hypo[j] = X[i][j]    
            elif hypo[j] != X[i][j]:
                hypo[j] = "?"        

print("Final Hypothesis:")
print(hypo)
```
