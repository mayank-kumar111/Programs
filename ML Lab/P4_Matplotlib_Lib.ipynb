```python
'''1. plt.plot()
Definition: Used to create a line plot by connecting data points with straight lines.

Syntax: plt.plot(x, y, format_string, **kwargs)

Example:'''
import matplotlib.pyplot as plt
'''x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, 'bo-')  # 'bo-' means blue color, circle markers, solid line
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()'''
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
'''2. plt.xlabel() and plt.ylabel()
Definition: Sets the labels for the x-axis and y-axis respectively.

Syntax: plt.xlabel("text") | plt.ylabel("text")

Example:'''
plt.plot([1, 2], [10, 20])
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.show()
```

```python
'''3. plt.title()
Definition: Adds a title to the top of the current plot.

Syntax: plt.title("Title Name")

Example:'''
plt.plot([1, 2], [5, 5])
plt.title("Constant Velocity Graph")
plt.show()
```

```python
'''4. plt.scatter()
Definition: Creates a scatter plot where individual data points are represented as dots without connecting lines.

Syntax: plt.scatter(x, y, color, marker)

Example:'''
plt.scatter([1, 2, 3, 4], [10, 15, 7, 12], color='red')
plt.show()
```

```python
'''5. plt.bar()
Definition: Creates a vertical bar chart to compare categories of data.

Syntax: plt.bar(x, height, width, color)

Example:'''
plt.bar(['A', 'B', 'C'], [10, 20, 15], color='green')
plt.show()
```

```python
'''6. plt.hist()
Definition: Computes and draws a histogram to show the frequency distribution of a dataset.

Syntax: plt.hist(data, bins)

Example:'''

data = [1, 2, 2, 3, 3, 3, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

```python
'''7. plt.pie()
Definition: Creates a circular pie chart to show proportions of a whole.

Syntax: plt.pie(sizes, labels, autopct)

Example:'''

plt.pie([30, 70], labels=['Task A', 'Task B'], autopct='%1.1f%%')
plt.show()
```

```python
'''8. plt.legend()
Definition: Adds a legend (key) to identify different lines or data series in a plot.

Syntax: plt.legend(loc="upper right")

Example:'''

plt.plot([1, 2], [1, 2], label='Growth')
plt.legend()
plt.show()
```

```python
'''9. plt.grid()
Definition: Adds grid lines (vertical and horizontal) to the background of the plot for easier reading.

Syntax: plt.grid(True)

Example:'''

plt.plot([1, 5], [1, 5])
plt.grid(True)
plt.show()
```

```python
'''10. plt.savefig()
Definition: Saves the current figure to a file (PNG, JPG, PDF, etc.) instead of just displaying it.

Syntax: plt.savefig("filename.png")

Example:'''

plt.plot([1, 2, 3], [1, 2, 3])
plt.savefig("my_plot.png")
```

```python
'''Category: Advanced Plotting
1. plt.stem()
Definition: Creates a stem plot, which is useful for visualizing discrete data points (like digital signals).

Syntax: plt.stem(x, y)

Example:'''

x = [0, 1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 2, 6]
plt.stem(x, y)
plt.show()
```

```python
'''2. plt.boxplot()
Definition: Draws a box-and-whisker plot to show the distribution of data based on a five-number summary (Minimum, First Quartile, Median, Third Quartile, and Maximum).

Syntax: plt.boxplot(data)

Example:'''

data = [10, 20, 22, 25, 30, 35, 50]
plt.boxplot(data)
plt.show()
```

```python
'''Category: Visual Customization
3. plt.grid()
Definition: Configures the visibility and appearance of grid lines in the background.

Syntax: plt.grid(color, linestyle, linewidth)

Example:'''

plt.plot([1, 2, 3], [10, 20, 30])
plt.grid(True, color='grey', linestyle='--')
plt.show()
```

```python
'''4. plt.axis()
Definition: Allows you to set the range of the x and y axes or change the scale (e.g., equal scaling).
Syntax: plt.axis([xmin, xmax, ymin, ymax]) or plt.axis('equal')
Example:'''

plt.plot([1, 2, 3], [1, 2, 3])
plt.axis([0, 5, 0, 10])
plt.show()
```

```python
'''5. plt.text()
Definition: Adds a text string at an arbitrary $(x, y)$ location on the plot.
Syntax: plt.text(x, y, "string")
Example:'''

plt.plot([1, 2, 3], [1, 4, 9])
plt.text(2, 4, "Peak Point")
plt.show()
```

```python
'''Category: Figure & Coordinate Management
6. plt.figure()
Definition: Used to create a new figure object or adjust the size of the plot.

Syntax: plt.figure(figsize=(width, height))

Example:'''

plt.figure(figsize=(10, 5))
plt.plot([1, 2], [1, 2])
plt.show()
```

```python
'''7. plt.xticks() and plt.yticks()
Definition: Sets the tick locations and labels on the x or y axis.

Syntax: plt.xticks(ticks, labels)

Example:'''

plt.plot([1, 2, 3], [10, 20, 30])
plt.xticks([1, 2, 3], ['Low', 'Med', 'High'])
plt.show()
```

```python
'''8. plt.tight_layout()
Definition: Automatically adjusts subplot parameters so that subplots fit into the figure area without overlapping.

Syntax: plt.tight_layout()

Example:'''

plt.subplot(2, 1, 1)
plt.title("Title 1")
plt.subplot(2, 1, 2)
plt.title("Title 2")
plt.tight_layout()
plt.show()
```

```python
'''Category: Specialized Visuals
9. plt.errorbar()
Definition: Plots data points with attached error bars to show uncertainty or variability.

Syntax: plt.errorbar(x, y, yerr=error_values)

Example:'''

x = [1, 2, 3]
y = [10, 20, 30]
error = [1, 2, 1.5]
plt.errorbar(x, y, yerr=error, fmt='o')
plt.show()
```

```python
'''10. plt.fill_between()
Definition: Fills the area between two horizontal curves with a specific color.

Syntax: plt.fill_between(x, y1, y2, color)

Example:'''

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.show()
```
