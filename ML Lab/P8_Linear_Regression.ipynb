```python
# Linear Regression using Matplotlib

import matplotlib.pyplot as plt

x_input = input("Enter X values (e.g., 1 2 3): ").split()
y_input = input("Enter Y values (e.g., 2 4 6): ").split()

x = [float(i) for i in x_input]
y = [float(i) for i in y_input]
n = len(x)

sum_x  = sum(x)
sum_y  = sum(y)
sum_xy = sum(val_x * val_y for val_x, val_y in zip(x, y))
sum_x2 = sum(val_x**2 for val_x in x)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
c = (sum_y - m * sum_x) / n

print(f"\nYour Equation: y = {m}x + {c}")

new_x = float(input("\nEnter a number for X to predict Y: "))
prediction = (m * new_x) + c
print(f"When X is {new_x}, Y will be about {prediction}")

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()
```

```python
# Linear Regression using NumPy
import numpy as np

print("Enter the data points for X (separated by space):")
x = np.array(list(map(float, input().split())))

print("Enter the data points for Y (separated by space):")
y = np.array(list(map(float, input().split())))

n = len(x)

if n != len(y):
    print("Error: X and Y must have the same number of data points.")
else:
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)
    sum_x2 = np.sum(x**2)

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)

    c = (sum_y - m * sum_x) / n

    print(f"\n--- Regression Results ---")
    print(f"Slope (m): {m:.4f}")
    print(f"Intercept (c): {c:.4f}")
    print(f"Regression Equation: y = {m:.4f}x + {c:.4f}")

    test_x = float(input("\nEnter a value of x to predict y: "))
    prediction = m * test_x + c
    print(f"Predicted y for x={test_x} is: {prediction:.4f}")
```

```python
# Linear Regression using NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as mtp
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

x_mean = np.mean(x)
y_mean = np.mean(y)
A = x - x_mean
B= y - y_mean
C = A ** 2
m = np.sum(A * B) / np.sum(C)
c = y_mean - m * x_mean
print("Slope:", m)
print("Intercept:", c)

mtp.plot(x,y,color="red")
mtp.xlabel('X')
mtp.ylabel('Y')
mtp.title('Linear Regression')
mtp.show()
```

```python
# Linear Regression using scikit-learn and evaluation metrics.
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("Linear_Regression.csv")
print(data.head())

# Separate the features (X) and target variable (y)
x = data.iloc[:, :-1].values # This selects all columns except the last one, which is 'Salary (y)'
y = data.iloc[:,1].values

from sklearn.model_selection import train_test_split
# x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state=0)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=1)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)
x_pred = regressor.predict(x_train)
# print(x_pred)
print(y_pred)

# Import evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
import numpy as np # np is already used for rmse, ensuring it's imported correctly

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

meap = mean_absolute_percentage_error(y_test, y_pred)
print("Mean Absolute Percentage Error (MAPE):", meap)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)

r2=r2_score(y_test,y_pred)
print("R2 Score:",r2)

accuracy_percent=r2*100
print(f"Accuracy: {accuracy_percent:.2f}%")

plt.scatter(x_train[:, 0], y_train, color="red")
plt.plot(x_train[:, 0], x_pred, color="blue")
plt.title("Years of Experience vs Salary (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

x1 = [[9.25]]
y1 = regressor.predict(x1)
print(f"Predicted salary for 9.25 years of experience: {y1[0]:.2f}")
```
